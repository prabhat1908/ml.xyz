```python
!pip install streamlit
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
st.set_page_config(page_title="KNN Weather Prediction", layout="centered")
st.title("üå¶Ô∏è KNN Weather Classification")
```

```python
# [Temperature, Humidity]
X = np.array([
    [30, 70],
    [25, 80],
    [27, 60],
    [31, 65],
    [23, 85],
    [28, 75]
])
```

```python
# 0 = Sunny, 1 = Rainy
y = np.array([0, 1, 0, 0, 1, 1])
```

```python
# -------------------- KNN Model --------------------
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)
```

```python

# -------------------- User Input --------------------
st.subheader("Enter Weather Conditions")

temp = st.slider("Temperature (¬∞C)", 20, 35, 26)
humidity = st.slider("Humidity (%)", 50, 90, 78)

new_point = np.array([[temp, humidity]])
prediction = knn.predict(new_point)[0]
```

```python
# -------------------- Plot --------------------
fig, ax = plt.subplots(figsize=(7, 5))

# Sunny points
ax.scatter(X[y == 0, 0], X[y == 0, 1], label="Sunny", s=100)

# Rainy points
ax.scatter(X[y == 1, 0], X[y == 1, 1], label="Rainy", s=100)
```

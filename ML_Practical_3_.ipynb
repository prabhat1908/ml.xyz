```python
import pandas as pd
import numpy as np
import math

data = pd.DataFrame({
    'Outlook':['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny','Sunny','Rain'],
    'Humidity':['High','High','High','High','Normal','Normal','Normal','High','Normal','High'],
    'PlayTennis':['No','No','Yes','Yes','Yes','No','Yes','No','Yes','Yes']
})

def entropy(col):
    probs = col.value_counts(normalize=True)
    return -sum(p*np.log2(p) for p in probs)

def information_gain(df, attr, target):
    total = entropy(df[target])
    vals = df[attr].unique()
    weighted = sum(
        (len(df[df[attr]==v])/len(df)) *
        entropy(df[df[attr]==v][target])
        for v in vals
    )
    return total-weighted

def id3(df, target, attrs):
    if len(df[target].unique())==1:
        return df[target].iloc[0]
    if len(attrs)==0:
        return df[target].mode()[0]

    gains = [information_gain(df,a,target) for a in attrs]
    best = attrs[np.argmax(gains)]

    tree={best:{}}
    for v in df[best].unique():
        subset=df[df[best]==v]
        remaining=[a for a in attrs if a!=best]
        tree[best][v]=id3(subset,target,remaining)
    return tree

tree=id3(data,'PlayTennis',['Outlook','Humidity'])
print(tree)

```

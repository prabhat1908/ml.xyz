```python
#Prabhat
import sys
!{sys.executable} -m pip install -q streamlit
import streamlit as st

tree = {
 'Outlook':{
  'Overcast':'Yes',
  'Rain':{'Humidity':{'High':'Yes','Normal':'No'}},
  'Sunny':{'Humidity':{'High':'No','Normal':'Yes'}}
 }
}

st.title("Play Tennis Prediction")

outlook = st.selectbox("Outlook",["Sunny","Overcast","Rain"])
humidity = st.selectbox("Humidity",["High","Normal"])

if st.button("Predict"):
    result = tree['Outlook'][outlook]
    if isinstance(result,dict):
        result = result['Humidity'][humidity]
    st.success(result)
```
